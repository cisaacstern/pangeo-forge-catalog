```python
import pystac
import fsspec
import xarray as xr
pystac.__version__  # make sure we have "pystac>=1.0.0"
```

```python
base = "https://raw.githubusercontent.com/cisaacstern/pangeo-forge-catalog/first-pass/stac"
keys = "/".join(["swot-adac", "GIGATL", "Region01", "surface", "fma"])
path = f"{base}/{keys}.json"
```

```python
item = pystac.read_file(path)
item
```

```python
asset = item.assets["zarr_osn"]
asset
```

```python
store = fsspec.get_mapper(asset.href, **asset.extra_fields["xarray:storage_options"])
ds = xr.open_zarr(store, **asset.extra_fields["xarray:open_kwargs"])
ds
```
